```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

### VM1 1/2CPU
### VM2 1/4CPU
### VM3 1/8CPU

```python
vm1cpu = 1/2
vm2cpu = 1/4
vm3cpu = 1/8
```

```python
N = [10, 100, 1000]
```

```python
lamda = 0.9
```

```python
#Decide type of vm
def getTypeVM():
    vmRan=random.random()
    cpu = 0
    if(vmRan <= 1/3):
        cpu = 1/2
    elif(vmRan<=2/3):
        cpu = 1/4
    else:
        cpu = 1/8
    return cpu
```

```python
def getNumberVMs(n):
    vmno=0
    for i in range(n):
        if(random.random() <= lamda):
            vmno+=1
    return vmno
```

```python
def checkdeparture(servers):
    for m in range(len(servers)):
        qLen = len(servers[m]['res'])
        for j in range(qLen):
            if(servers[m]['res'][j] == vm1cpu):
                if(random.random() <= 1/10 ):
                    del servers[m]['res'][j]
                    del servers[m]['slot'][j]
                    break
            elif(servers[m]['res'][j] == vm2cpu):
                if(random.random() <= 1/8 ):
                    del servers[m]['res'][j]
                    del servers[m]['slot'][j]
                    break
            elif(servers[m]['res'][j] == vm3cpu):
                if(random.random() <= 1/30 ):
                    del servers[m]['res'][j]
                    del servers[m]['slot'][j]
                    break
```

```python
sever_num = []
```

```python
def getONServers(servers):
    server_count = 0
    for s in servers:
        if(len(s['res'])>0):
            server_count += 1
    return server_count
```

# First Fit

```python
servers = []
for i in range(10**2):
    checkdeparture(servers)
    for k in range(getNumberVMs(1000)):
        requestedRes = getTypeVM()
        serverLen = len(servers)
        if(serverLen == 0):
            servers.append({'res' : [requestedRes],'slot' : [i]})
        else:
            found = False
            for j in range(serverLen):
                res = np.sum(servers[j]['res'])+requestedRes
                if(res <= 1.0):
                    servers[j]['res'].append(requestedRes)
                    servers[j]['slot'].append(i)
                    found = True
                    break
            if(not found):
                servers.append({'res': [requestedRes],'slot': [i]})
    
    server_count = 0
    for s in servers:
        if(len(s['res'])>0):
            server_count += 1
    sever_num.append(server_count)
```
